```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
  if df[i].dtype == 'object':
    df[i]=l.fit_transform(df[i])
```

```python
df.shape
```

```python
x = df.drop('DALYs (Disability-Adjusted Life Years) - Mental disorders',axis = 1)
y = df['DALYs (Disability-Adjusted Life Years) - Mental disorders']
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.20, random_state=2)
```

```python
print("xtrain: ", xtrain.shape)
print("xtest: ", xtest.shape)
print("ytrain: ", ytrain.shape)
print("ytest: ", ytest.shape)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr = LinearRegression()
lr.fit(xtrain,ytrain)
ytrain_pred = lr.predict(xtrain)
mse_lr = mean_squared_error(ytrain, ytrain_pred)
rmse_lr = (np.sqrt(mean_squared_error(ytrain,ytrain_pred)))
r2_lr = r2_score(ytrain,ytrain_pred)
print("The Linear Regression model performance for training set")
print("----------------------------------------")
print('MSE is {}'.format(mse_lr))
print('RMSE is {}'.format(rmse_lr))
print('R2 score is {}'.format(r2_lr))
```

Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain,ytrain)
ytrain_pred = rf.predict(xtrain)
mse_rf = mean_squared_error(ytrain, ytrain_pred)
rmse_rf = (np.sqrt(mean_squared_error(ytrain,ytrain_pred)))
r2_rf = r2_score(ytrain,ytrain_pred)
print("The Random Forest Regression model performance for training set")
print("----------------------------------------")
print('MSE is {}'.format(mse_rf))
print('RMSE is {}'.format(rmse_rf))
print('R2 score is {}'.format(r2_rf))
```

Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
lasso = Lasso(alpha=0.1)
lasso.fit(xtrain, ytrain)
y_pred = lasso.predict(xtrain)
mse_lasso = mean_squared_error(ytrain, ytrain_pred)
rmse_lasso = (np.sqrt(mean_squared_error(ytrain,ytrain_pred)))
r2_lasso = r2_score(ytrain,ytrain_pred)
print("The Lasso Regression model performance for training set")
print("----------------------------------------")
print('MSE is {}'.format(mse_lasso))
print('RMSE is {}'.format(rmse_lasso))
print('R2 score is {}'.format(r2_lasso))
```

Decision Tree Regresssor

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(xtrain,ytrain)
predict3=dtr.predict(xtest)
mse_dtr=mean_squared_error(ytest, predict3)
rmse_dtr=(np.sqrt(mean_squared_error(ytest, predict3)))
r2_dtr=r2_score(ytest, predict3)
print("The Decision Tree Regression model performance for training set")
print("----------------------------------------")
print('MSE is ',mse_dtr)
print('RMSE is ',rmse_dtr)
print('R2 score is ',r2_dtr)
```
